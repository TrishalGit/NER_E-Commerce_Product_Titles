```python
import csv
import os
import re 

directory = os.getcwd()

with open(directory + '/eBay_ML_Challenge_Dataset_2022/Train_Tagged_Titles.tsv', mode ='r') as file:
    csvFile = csv.reader(file, delimiter="\t")
    c = 0
    train = []
    dev = []
    test = []
    prev_tag = 'O'
    record = '1'
    for lines in csvFile:
#         print(lines)
        c += 1
        if c == 1:
            continue
        if lines[-1] == 'No Tag' or lines[-1] == 'Obscure':
            tag = 'O'
        elif lines[-1] == '':
            tag = prev_tag
        else:
            tag = re.sub('[^a-zA-Z0-9 \n\.]', '', lines[-1])
            tag = tag.replace(" ", "")
            tag = "I-" + tag
            
        if c < 6000:
            if record != lines[0]:
                train.append('')
            train.append(lines[-2] + ' ' + tag)
        elif c < 8000:
            if record != lines[0]:
                dev.append('')
            dev.append(lines[-2] + ' ' + tag)
        else:
            if record != lines[0]:
                test.append('')
            test.append(lines[-2] + ' ' + tag)
        prev_tag = tag
        record = lines[0]
        if(c == 10000):
            break
with open(directory + '/eBay_ML_Challenge_Dataset_2022/eng.train', 'w') as the_file:
    for i in range(len(train)):
        the_file.write(train[i])
        the_file.write('\n')   
        
with open(directory + '/eBay_ML_Challenge_Dataset_2022/eng.testa', 'w') as the_file:
    for i in range(len(dev)):
        the_file.write(dev[i])
        the_file.write('\n') 

with open(directory + '/eBay_ML_Challenge_Dataset_2022/eng.testb', 'w') as the_file:
    for i in range(len(test)):
        the_file.write(test[i])
        the_file.write('\n') 
print(train)
```
