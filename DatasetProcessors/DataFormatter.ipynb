```python
import csv
import os
import re 

directory = os.getcwd()

with open(directory + '/Data/trainMixed.csv', mode ='r') as file:
    csvFile = csv.reader(file, delimiter=",")
    c = 0
    train = []
    dev = []
    test = []
    prev_tag = 'O'
    record = '1'
    linesList = []

    mappedTags = []
    recNo = 0
    cc = 0
    for lines in csvFile:
        print(lines)
        print(len(lines), lines[-1])
        if lines[0] != recNo:
            recNo = lines[0]
            cc += 1
        c += 1
        if c == 1:
            continue
        if len(lines) == 3:
            if prev_tag == 'O':
                tag = prev_tag
            else:
                tag = prev_tag
                index = 0
                new_character = 'I'
                tag = tag[:index] + new_character + tag[index+1:]
        elif lines[-1] == 'No Tag' or lines[-1] == 'Obscure':
            tag = 'O'
        else:
            tag = re.sub('[^a-zA-Z0-9 \n\.]', '', lines[-1])
            tag = tag.replace(" ", "")
            mappedTags.append(lines[-1] + '~' + tag)
            tag = "B-" + tag
        if record != lines[0]:
            train.append('')
        if len(lines) == 3:
            train.append(lines[-1] + ' ' + tag)
        else:
            train.append(lines[-2] + ' ' + tag)
        prev_tag = tag
        record = lines[0]
        
with open(directory + '/Data/eng.trainMixed', 'w') as the_file:
    for i in range(len(train)):
        the_file.write(train[i])
        the_file.write('\n')   
print(len(train))
li = set(mappedTags)
# print(li)
alltagMappings = []
for i in li:
    words = i.split('~')
    alltagMappings.append([words[1], words[0]])
print(alltagMappings)
f = open(directory + '/Data/tagsMixed.csv', 'w', newline='')
w = csv.writer(f, delimiter=",")
w.writerows(alltagMappings)
f.close()
```
